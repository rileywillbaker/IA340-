```python
import warnings
warnings.simplefilter('ignore')

import pandas as pd

# Load your dataset from S3
df = pd.read_csv('s3://ia340-2025-fall-rb11/Electric_Vehicle_Population_Data.csv')

# Show the first 5 rows
df.head()

# Number of rows and columns
df.shape  

# Info about columns and data types
df.info()

# Summary statistics
df.describe()

```

```python
# Rename a column (example: change 'VIN (1-10)' to 'VIN')
df.rename(columns={'VIN (1-10)': 'VIN'}, inplace=True)

# Confirm the change
df.columns
```

```python
# Count missing values in each column
df.isnull().sum()
```

```python
# Drop rows with missing values
df_clean = df.dropna()

# Check again
df_clean.isnull().sum()
```

## Step C: Manipulate the Data

- I selected the **Make** column to see the first 10 manufacturers in the dataset.  
- I sorted the dataset by **Model Year**, which showed the newest vehicles at the top.  
- I sliced rows 510 to preview a subset of the dataset.  
- I filtered the dataset to only include **Battery Electric Vehicles (BEV)**, which gave me a smaller, focused dataset of fully electric vehicles.


```python
# Example: count vehicles by Model Year
year_counts = df_clean.groupby('Model Year').size()

# Show the grouped summary
year_counts.head()
```

```python
import matplotlib.pyplot as plt

# Line plot of vehicle counts by Model Year
year_counts.plot(kind='line', figsize=(8,5), title='Number of EVs by Model Year')
plt.xlabel('Model Year')
plt.ylabel('Count of Vehicles')
plt.show()
```

```python
# Bar chart of top 10 car makes
top_makes = df_clean['Make'].value_counts().head(10)

top_makes.plot(kind='bar', figsize=(10,5), title='Top 10 EV Manufacturers')
plt.xlabel('Car Make')
plt.ylabel('Number of Vehicles')
plt.show()
```

```python
# Histogram of electric range
df_clean['Electric Range'].plot(kind='hist', bins=30, figsize=(8,5), title='Distribution of Electric Range')
plt.xlabel('Electric Range (miles)')
plt.show()
```

## Step D: Group & Visualize the Data

- The **groupby by Model Year** shows how EV adoption has increased over time. The line plot clearly shows a rising trend in newer years.  
- The **bar chart of top 10 manufacturers** shows that Tesla, Nissan, and Chevrolet are among the leaders in EV registrations.  
- The **histogram of electric range** shows that most EVs fall into shorter ranges (e.g., <150 miles), but a smaller group of vehicles achieve much higher ranges.
